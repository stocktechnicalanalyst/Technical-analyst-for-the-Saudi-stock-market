```python
import pandas as pd                                               #to read excel files and deal with dataframes
import numpy as np                                                #to deal with arrays 
from sklearn.model_selection import train_test_split              #to split data
from sklearn.preprocessing import LabelEncoder                    #to encode data 
from sklearn import preprocessing
import matplotlib.pyplot as plt                                   #to plot data
import seaborn as sns                                             #to plot data
#from sklearn.preprocessing import Normalizer                      #to normalize data
```

```python
data = pd.read_csv('sp500_companies - Copy.csv', index_col=0)
data = data.loc[:, data.columns.intersection(['Country','Sector','Currentprice'])]
data = data[data['Sector'].isin(['Technology','Industrials','Financial Services'])]
data = data[data['Country'].isin(['United States','Ireland','Switzerland'])]
data.head(10)
```

```python
#plot categories 
sns.countplot(data['Country'])
plt.xlabel('Label')
plt.title('Number of each Companies in each Country')
plt.savefig('img/num_companies_country',dpi=300)
```

```python
sns.countplot(data['Sector'])
plt.xlabel('Label')
plt.title('Number of each Companies in each Sector')
plt.savefig('img/num_companies_sector',dpi=300)
```

```python
#display unique values to each column
print("Sector:")
print(np.unique(data['Sector']))
print('--------------------------------------------------------')
print("Country:")
print(np.unique(data['Country']))
print('--------------------------------------------------------')
```

```python
#encode data (from string to numerical values )
data['Sector'] = LabelEncoder().fit_transform(data['Sector'])
print(np.unique(data['Sector']))
print('---------------------------------------------------')
data['Country'] = LabelEncoder().fit_transform(data['Country'])
print(np.unique(data['Country']))
print('---------------------------------------------------')
```

```python
data.head(5)
```

```python
#split the the target and the features 

X =data.drop("Sector",axis = 1 )
Y = data['Sector']
X
```

```python
Y
```

```python
scaler = preprocessing.MinMaxScaler()
names = X.columns
d = scaler.fit_transform(X)
X = pd.DataFrame(d, columns=names)
X.head(5)
```

```python
#Splitting data

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=44, shuffle =True)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
import pandas as pd
import numpy as np
from tensorflow.python import keras
import tensorflow.python.keras as k
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Dropout
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from tensorflow.python.keras.utils import np_utils

data = pd.read_csv('sp500_companies - Copy.csv', index_col=0)
data = data.loc[:, data.columns.intersection(['Country','Sector','Currentprice'])]
data = data[data['Sector'].isin(['Technology','Industrials','Financial Services'])]
data = data[data['Country'].isin(['United States','Ireland','Switzerland'])]
data['Country'] = LabelEncoder().fit_transform(data['Country'])
X =data.drop("Sector",axis = 1 )
Y = data['Sector']

print(Y)
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
uniques, ids = np.unique(encoded_Y, return_inverse=True)
# convert integers to dummy variables (i.e. one hot encoded)
Y = np_utils.to_categorical(ids, len(uniques))


scaler = preprocessing.MinMaxScaler()
names = X.columns
d = scaler.fit_transform(X)
X = pd.DataFrame(d, columns=names)
Y
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, Y, test_size=0.25, random_state=44, shuffle =True)
```

```python
# Build the model
model = Sequential()
model.add(Dense(120, input_dim=2, activation='relu'))
model.add(Dropout(0.001))
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.001))
model.add(Dense(3, activation='softmax'))


# Configure a model for mean-squared error regression.
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpointer = k.callbacks.ModelCheckpoint(filepath="weights.h5", verbose=1, save_best_only=True)
```

```python
history = model.fit(X_train1, y_train1, epochs=5000, batch_size=32, verbose=1, validation_split=0.2, callbacks=[checkpointer])
model.summary()
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.savefig('img/NN_loss',dpi=300)
plt.show()
```

```python
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.savefig('img/NN_ACC',dpi=300)
plt.show()
```

```python
loss, accuracy = model.evaluate(X_train1, y_train1)
ANN_Acc = round(accuracy*100,2)
print(model.metrics_names[1],": ", round(accuracy*100,2),"%")
print(model.metrics_names[0],": ", loss)

print("------------------")
# Predict

print(model.predict(X_test1[:5]))
print("------------------")
print(y_test1[:5])
print("------------------")
for x in range(1,6):
    if(uniques[y_test1.argmax(1)][x] == 0):
        print("Financial Services")
    elif(uniques[y_test1.argmax(1)][x] == 1):
        print("Industrials")
    elif(uniques[y_test1.argmax(1)][x] == 2):
        print("Technology")    
```
